# Support Vector Machine Algorithm

## Import Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.append("../")

from src.svm import SVC
```

## Classification Case

```python
X = np.array([[0, 0], [1, 0], [1, 1], [0, 1]])
y = np.array([1, 1, -1, 1])

# Create object
clf = SVC(C=10)
clf.fit(X, y)

# Print
print(clf.coef)
print(clf.intercept)

# Visualize the decision boundary
# Create the decision boundary
w1, w2 = clf.coef
w0 = clf.intercept
m = -w1/w2
c = -w0/w2

# Create prediction
data = pd.DataFrame(X,
                    columns=["x1", "x2"])
data["y"] = y
data["y_pred"] = clf.predict(X)

# PLOT
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 10))

sns.scatterplot(data = data,
                x = "x1",
                y = "x2",
                hue = "y_pred",
                s = 200,
                ax = ax)

# Plot decision
x_ = np.linspace(-1, 2, 101)
y_ = m*x_ + c
ax.plot(x_, y_, "--r", alpha=0.5, label="decision boundary")

plt.xlim(-1, 2)
plt.ylim(-1, 2)
plt.legend()
plt.grid()
plt.show()
```
