## Decomposition

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import sys
sys.path.append("../")


from src.decomposition import PCA
from src.decomposition import SVD
```

```python
# Generate Data
# DATA 1
X = np.array([[-1, -1],
              [-2, -1],
              [-3, -2],
              [1, 1],
              [2, 1],
              [3, 2]])

# Plot
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 7))
sns.scatterplot(x=X[:,0], y=X[:,1], ax=ax, s=100, label="Original Data")

ax.set_xlabel("X_1", fontsize=16)
ax.set_ylabel("X_2", fontsize=16)
plt.show()
```

```python
# Fit with PCA
pca = PCA(n_components = 1)
pca.fit(X)
X_transformed = pca.transform(X)
X_inverse_transformed = pca.inverse_transform(X_transformed)

print(f"PCA RESULTS")
print("------------")
print("Explained Variance Ratio:")
print(pca.explained_variance_ratio_)
print("")
print("The principal components:")
print(pca.components_)
print("")
print("Transform original data to PC axis:")
print(X_transformed)
print("")
print("Transform back the PC data to original data:")
print(np.round(X_inverse_transformed, 2))
print("")

# Plot
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 7))

sns.scatterplot(x=X[:,0], y=X[:,1], ax=ax, s=100, label="Original Data")
sns.scatterplot(x = X_inverse_transformed[:,0],
                y = X_inverse_transformed[:,1],
                color="red", s=100, ax=ax, 
                label="Inverse Transform PCA")

ax.set_xlabel("X_1", fontsize=16)
ax.set_ylabel("X_2", fontsize=16)
ax.set_title("PCA", fontsize=25)
plt.legend()
plt.show()


```

```python
# Fit with SVD
#svd = TruncatedSVD(n_components=1)
svd = SVD(n_components=1)
svd.fit(X)
X_transformed = svd.transform(X)
X_inverse_transformed = svd.inverse_transform(X_transformed)

print(f"SVD RESULTS")
print("------------")
print("Explained Variance Ratio:")
print(svd.explained_variance_ratio_)
print("")
print("The principal components:")
print(svd.components_)
print("")
print("Transform original data to PC axis:")
print(X_transformed)
print("")
print("Transform back the PC data to original data:")
print(np.round(X_inverse_transformed, 2))
print("")


# PLOT
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 7))

sns.scatterplot(x=X[:,0], y=X[:,1], ax=ax, s=100, label="Original Data")
sns.scatterplot(x = X_inverse_transformed[:,0],
                y = X_inverse_transformed[:,1],
                color="red", s=100, ax=ax, 
                label="Inverse Transform SVD")

ax.set_xlabel("X_1", fontsize=16)
ax.set_ylabel("X_2", fontsize=16)
ax.set_title("SVD", fontsize=25)
plt.legend()
plt.show()
```
